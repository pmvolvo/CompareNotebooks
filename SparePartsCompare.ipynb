```python
import re
def regexCorrection(string):
    return re.sub('[^a-zA-Z0-9]', '', string.upper())
```

```python
def findDrawingNr(string):
    checkDrawing = re.compile(r'D\d\d\-')
    return checkDrawing.search(string) != None
```

```python
def stripSymbols(string):
    string = string.replace('-','')
    string = string.replace('_',' ')
    return string
    
```
